We can see that Dataset is not balance Clothing And Accessories is half of Household

```python
balanced_data = data.groupby('Category', group_keys=False).apply(lambda x: x.sample(n=min(8670, len(x)), random_state=2024))
```

```python
balanced_data.Category.value_counts()
```

```python
balanced_data.head()
```

```python
balanced_data.shape
```

```python
ok={
    'Books':0,
    'Clothing & Accessories':1,
    'Electronics':2,
    'Household':3,
}
```

```python
balanced_data['Category']=balanced_data['Category'].map(ok)
```

```python
balanced_data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(balanced_data['Product_Description'], balanced_data['Category'], test_size=0.2, random_state=202)
```

```python
X_train
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
v=CountVectorizer()
X_train_count=v.fit_transform(X_train.values)
X_train_count.toarray()[:2]
model=MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
x_test_t=v.transform(X_test)
```

```python
pred=model.predict(x_test_t)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
classf=classification_report(y_test,pred)
print(classf)
```
